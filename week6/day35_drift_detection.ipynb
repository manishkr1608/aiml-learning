```python
import pandas as pd
from sklearn.metrics import mean_squared_error
from scipy.stats import ks_2samp
import numpy as np
```

```python
# Simulate training and live data
train = pd.DataFrame({"feature": np.random.normal(0, 1, 500)})
live = pd.DataFrame({"feature": np.random.normal(0.3, 1.2, 500)})
```

```python
# Simple statistical drift check (Kolmogorov–Smirnov test)
ks_stat, p_value = ks_2samp(train["feature"], live["feature"])
print(f"KS statistic={ks_stat:.3f}, p={p_value:.5f}")
```

```python
if p_value < 0.05:
    print("⚠️ Drift detected in feature distribution!")
else:
    print("✅ No significant drift detected.")
```
