# Day 4: Exploratory Data Analysis basics
- Load Sample Dataset
- Handle Missing Values
- Find Correlation
- Plot histogram, Scatter Plot and Heat Map
 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
df = pd.DataFrame({
    "age": [22, 25, 29, 35, 40, 28, 50, 32],
    "income": [30, 45, 50, 80, 90, 55, 100, 70],
    "purchased": [0, 1, 0, 1, 1, 0, 1, 1]
})
```

```python
print(df.info())
print(df.describe())
```

```python
print("Missing values:\n", df.isnull().sum())
```

```python
print("Correlation:\n", df.corr())
```

```python
sns.histplot(df["age"], bins=5, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.scatterplot(x="age", y="income", hue="purchased", data=df)
plt.title("Income vs Age (colored by Purchase)")
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```
