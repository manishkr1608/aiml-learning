```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "date": ["2023-01-01","2023-01-01","2023-01-02","2023-01-02","2023-01-03"],
    "product": ["A","B","A","B","A"],
    "sales": [100,200,150,250,175]
}
df = pd.DataFrame(data)

# totals per product
totals = df.groupby("product")["sales"].sum()
print(totals)

# daily avg sales for product A
avg_a = df[df["product"]=="A"]["sales"].mean()
print("Avg sales (A):", avg_a)

# quick chart
totals.plot(kind="bar", title="Total Sales by Product")
plt.xlabel("Product"); plt.ylabel("Sales"); plt.tight_layout(); plt.show()
```
